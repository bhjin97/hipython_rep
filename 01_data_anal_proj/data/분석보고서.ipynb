```python
import seaborn as sns

# 항공사 리스트
airlines = df['airline'].unique()

# 지연 유형 컬럼들
delay_cols = ['pct_carrier_delay', 'pct_atc_delay', 'pct_weather_delay']


plt.figure(figsize=(18, 5))

# 지연유형별 subplot
for i, col in enumerate(delay_cols):
    plt.subplot(1, 3, i + 1)
    for airline in airlines:
        sns.histplot(
            data=df[df['airline'] == airline],
            x=col,
            label=airline,
            kde=False,
            bins=20,
            alpha=0.25
        )
    plt.title(f'{col} by Airline')
    plt.xlabel('Delay')
    plt.ylabel('Frequency')
    plt.legend()

plt.tight_layout()
plt.show()

```

```python

# 지연 유형 컬럼들
delay_cols = ['pct_carrier_delay', 'pct_atc_delay', 'pct_weather_delay']

# 데이터를 긴 형식으로 변환 (멜팅)
# 'airline'은 그대로 두고, delay_cols에 있는 컬럼들을 'delay_type'과 'percentage'로 변환
df_melted = df.melt(id_vars=['airline'],
                     value_vars=delay_cols,
                     var_name='delay_type',
                     value_name='percentage')

# 'delay_type'의 라벨을 더 읽기 쉽게 변경
df_melted['delay_type'] = df_melted['delay_type'].replace({
    'pct_carrier_delay': 'Carrier Delay',
    'pct_atc_delay': 'ATC Delay',
    'pct_weather_delay': 'Weather Delay'
})

# catplot을 사용하여 시각화
# col='airline'로 항공사별로 서브플롯을 생성
# x='delay_type'으로 각 지연 유형을 x축에 표시
# y='percentage'로 지연 비율을 y축에 표시
# kind='bar'로 막대 그래프 선택
# col_wrap=3으로 한 행에 3개의 서브플롯을 표시
# height와 aspect로 각 서브플롯의 크기 조절
g = sns.catplot(
    data=df_melted,
    x='delay_type',
    y='percentage',
    col='airline',
    kind='bar',
    col_wrap=3, # 한 줄에 3개의 서브플롯
    height=4, # 각 서브플롯의 높이
    aspect=1.2, # 각 서브플롯의 너비 비율
    errorbar=None, # 오차 막대 제거 (필요시 'sd' 또는 'se'로 변경)
    palette='viridis' # 색상 팔레트 변경
)

# 각 서브플롯의 제목 설정
g.set_titles("{col_name}")
g.set_axis_labels("Delay Type", "Average Delay Percentage")
plt.suptitle('Average Delay Percentage by Delay Type for Each Airline', y=1.02, fontsize=16) # 전체 제목 추가

plt.tight_layout(rect=[0, 0, 1, 0.98]) # 전체 제목을 위한 여백 조정
plt.show()
```

3. 추가 분석

```python
# 항공사 리스트 (실제 df에 있는 unique 값 사용)
airlines = df['airline'].unique().tolist()
# airlines = ['Alaska', 'American', 'Delta', 'Jet Blue', 'Southwest', 'United']

# 지연 유형 컬럼들 
delay_cols = ['pct_carrier_delay', 'pct_atc_delay', 'pct_weather_delay']
delay_labels = ['Carrier Delay', 'ATC Delay', 'Weather Delay'] # 파이 그래프 라벨용

n_airlines = len(airlines)
n_rows = (n_airlines + 2) // 3 # 3개씩 한 줄에 배치
n_cols = 3

plt.figure(figsize=(n_cols * 6, n_rows * 5)) # 전체 figure 크기 조정
plt.suptitle('Delay Cause Breakdown by Airline', y=1.02, fontsize=16) # 전체 제목

# 항공사별 파이 그래프 생성
for i, airline in enumerate(airlines):
    plt.subplot(n_rows, n_cols, i + 1) # i+1은 현재 서브플롯 인덱스

    # 해당 항공사의 지연 유형별 평균 계산
    airline_data = df[df['airline'] == airline][delay_cols].mean()

    # 모든 지연 유형의 합이 0인 경우 (즉, 해당 항공사의 지연 데이터가 없거나 모두 0인 경우)
    if airline_data.sum() == 0:
        plt.text(0.5, 0.5, 'No Significant Delay Data',
                 horizontalalignment='center', verticalalignment='center',
                 transform=plt.gca().transAxes, fontsize=12, color='gray')
        plt.title(f'{airline} Delay Breakdown')
        continue

    # 파이 그래프 그리기
    plt.pie(
        airline_data,
        labels=delay_labels,
        autopct='%1.1f%%',
        startangle=90,
        pctdistance=0.85
    )
    plt.title(f'{airline} Delay Breakdown')

plt.tight_layout(rect=[0, 0, 1, 0.98]) # 전체 제목을 위한 여백 조정
plt.show()
```

```python
# 항공사 리스트
airlines = df['airline'].unique().tolist()

# 지연 유형 컬럼들
delay_cols = ['pct_carrier_delay', 'pct_atc_delay', 'pct_weather_delay']

# 서브플롯 그리드 설정 (항공사 수에 따라 조정)
n_airlines = len(airlines)
# 각 항공사당 2개의 산점도 (carrier vs atc, carrier vs weather)를 그릴 것이므로 열을 2배로 설정
n_rows = n_airlines
n_cols_per_airline = 2 # 각 항공사당 2개의 그래프
total_cols = n_cols_per_airline

plt.figure(figsize=(total_cols * 7, n_rows * 5)) # 전체 figure 크기 조정
plt.suptitle('Delay Type Scatter Plots by Airline', y=1.02, fontsize=18) # 전체 제목

# 항공사별 서브플롯 생성
for i, airline in enumerate(airlines):
    # 해당 항공사의 데이터 필터링
    airline_df = df[df['airline'] == airline].copy() # SettingWithCopyWarning 방지를 위해 .copy() 사용

    # 첫 번째 산점도: Carrier Delay vs ATC Delay
    plt.subplot(n_rows, total_cols, i * total_cols + 1) # (i * 2) + 1 번째 서브플롯
    sns.scatterplot(
        data=airline_df,
        x='pct_carrier_delay',
        y='pct_atc_delay',
        alpha=0.3, # 점들의 투명도 조절
        s=10 # 점 크기 조절
    )
    plt.title(f'{airline}: Carrier Delay vs ATC Delay')
    plt.xlabel('Carrier Delay (%)')
    plt.ylabel('ATC Delay (%)')
    plt.grid(True, linestyle='--', alpha=0.6) # 그리드 추가

    # 두 번째 산점도: Carrier Delay vs Weather Delay
    plt.subplot(n_rows, total_cols, i * total_cols + 2) # (i * 2) + 2 번째 서브플롯
    sns.scatterplot(
        data=airline_df,
        x='pct_carrier_delay',
        y='pct_weather_delay',
        alpha=0.3, # 점들의 투명도 조절
        s=10 # 점 크기 조절
    )
    plt.title(f'{airline}: Carrier Delay vs Weather Delay')
    plt.xlabel('Carrier Delay (%)')
    plt.ylabel('Weather Delay (%)')
    plt.grid(True, linestyle='--', alpha=0.6) # 그리드 추가

plt.tight_layout(rect=[0, 0, 1, 0.98]) # 전체 제목을 위한 여백 조정
plt.show()
```

5. 공분산 확인 - 전체 비교, 항공사별 비교

```python
# 지연율 열만 선택
delay_cols = ['pct_carrier_delay', 'pct_atc_delay', 'pct_weather_delay']
# 공분산 행렬
cov_matrix = df[delay_cols].cov()
print(cov_matrix)
```

```python
sns.heatmap(cov_matrix, annot=True, fmt=".2f", cmap='Blues')
plt.title("지연 유형 간 공분산 행렬")
plt.show()
```

```python
import numpy as np
# 항공사 리스트
airlines = df['airline'].unique().tolist()

# 지연 유형 컬럼들
delay_cols = ['pct_carrier_delay', 'pct_atc_delay', 'pct_weather_delay']
delay_labels = ['Carrier', 'ATC', 'Weather'] # 히트맵 라벨용

# 서브플롯 그리드 설정 (항공사 수에 따라 조정)
n_airlines = len(airlines)
n_rows = (n_airlines + 2) // 3 # 3개씩 한 줄에 배치
n_cols = 3

plt.figure(figsize=(n_cols * 6, n_rows * 5)) # 전체 figure 크기 조정
plt.suptitle('Covariance of Delay Types by Airline', y=1.02, fontsize=16, fontweight='bold') # 전체 제목

# 항공사별 공분산 히트맵 생성
for i, airline in enumerate(airlines):
    plt.subplot(n_rows, n_cols, i + 1) # i+1은 현재 서브플롯 인덱스

    # 해당 항공사의 데이터 필터링
    airline_df = df[df['airline'] == airline].copy()

    # 지연 유형 컬럼만 선택하여 공분산 행렬 계산
    covariance_matrix = airline_df[delay_cols].cov()

    # 히트맵 그리기
    sns.heatmap(
        covariance_matrix,
        annot=True,     # 각 셀에 값 표시
        fmt=".1f",      # 값의 소수점 첫째 자리까지 표시
        cmap='coolwarm',# 색상 팔레트 설정 (양수/음수 구별에 좋음)
        linewidths=.5,  # 셀 간 경계선
        linecolor='black', # 경계선 색상
        square=True,    # 셀을 정사각형으로 만듦
        cbar_kws={'label': 'Covariance'} # 컬러 바 라벨
    )
    plt.title(f'{airline} Covariance', fontsize=12)
    plt.xticks(ticks=np.arange(len(delay_labels)) + 0.5, labels=delay_labels, rotation=45, ha='right')
    plt.yticks(ticks=np.arange(len(delay_labels)) + 0.5, labels=delay_labels, rotation=0)

plt.tight_layout(rect=[0, 0, 1, 0.98]) # 전체 제목을 위한 여백 조정
plt.show()
```

## 인사이트 도출
다음은 각 지연 유형별 지연률이 가장 높은 항공사이다
- 항공사 자체 딜레이: 아메리칸 항공사
- 관제로 인한 딜레이: 젯블루 항공사
- 기상으로 인한 딜레이: 아메리칸    

다음은 각 항공사 별로 어떤 지연요인이 지연횟수에 더 큰 영향을 미치는지를 보여준다
-  델타항공: 날씨로 인한 지연
- 아메리칸 항공: 항공사,날씨로 인한 지연
- 알래스카 항공: 날씨로 인한 지연
- 젯블루 항공: 항공사,날씨로 인한 지연
- 사우스웨스트 항공: 날씨로 인한 지연
- 유나이티드 항공: 항공사,날씨로 인한 지연   

전체적으로 봤을 때는 각 지연 유형별로는 서로 거의 영향을 주지 않는다는 결과가 나오고
항공사별로 나누어서 지연유형을 비교해보아도 서로 영향을 주지 않는다는 결과를 얻었다


## 결론 및 제언
위와 같은 결과에서 알 수 있듯이 대부분의 항공사에서 세가지 지연유형의 비율은 크게 차이 나지 않았지만    
날씨와 항공사 자체의 잘못으로 인한 지연 케이스가 가장 많았다는 것을 알 수 있었다
   
하지만 지연횟수와 별개로 지연율에 대한 결과를 보면 날씨보다는 항공사와 관제로 인한 지연율이 훨씬 높다는 것을 볼 수 있다

따라서 지연요인중에서 날씨보다는 항공사와 관제센터의 비중이 크다고 볼 수 있겠다
    
정책적 개선안: 항공사 책임으로 인한 지연에 대한 처벌을 강화하되 항공사별로 지연되는 원인을 파악하여 항공사 내부의 시스템 개선에 도움이 되는 정책도 필요해보임
