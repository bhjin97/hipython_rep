```python
dates = ['2024-10-01', '2024-10-02', '2024-10-03', '2024-10-04', '2024-10-05']
prices = [100, 102, 101, 105, 103]
```

```python
plt.Figure(figsize=(10,6))
plt.title('stock prices over 5 days')
plt.xlabel('Date')
plt.ylabel('Price')
plt.ylim(95,110)
plt.grid()

plt.plot(dates, prices,marker='o',label='stock prices')
plt.legend(loc='upper right')
```

```python
plt.plot(dates, prices)
plt.plot(dates, prices)
```

```python
plt. subplot(1,2,1)
plt.title('stock prices over 5 days')
plt.plot(dates, prices)# line 차트
plt.xticks(rotation=40)

plt. subplot(1,2,2)
plt.title('stock prices over 5 days')
plt.bar(dates, prices) # bar차트
plt.xticks(rotation=40)

plt.savefig('my_stock_prices_chart.png')
```
