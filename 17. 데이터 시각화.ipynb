# 새로운 데이터 셋

```python
tips_df=sns.load_dataset('tips')
tips_df.head()
```

## 1변량 그래프 :  히스토그램

```python
plt.hist(data=tips_df, x= 'total_bill', bins=20)
```

```python
plt.subplot(121)
plt.hist(data=tips_df, x= 'total_bill', bins=20)
plt.subplot(122)
plt.hist(data=tips_df, x= 'tip', bins=20, color='y')
```

```python
fig=plt.figure(figsize=[12,5])
axes1=fig.add_subplot(1,2,1) #total_bill
axes2=fig.add_subplot(1,2,2) #tips
axes1.hist(data=tips_df, x='total_bill', bins=20)
axes2.hist(data=tips_df, x='tip', bins=20, color='m')
plt.tight_layout()
plt.show()
```

```python
# boxplot
fig=plt.figure(figsize=[12,5])
axes=fig.add_subplot(111) #total_bill
axes.boxplot(data=tips_df, x='total_bill')
plt.show()
# 양 끝은 최소,최대값 / 중간 노란선은 중앙값 / 위에는 아상치
```

```python
fig=plt.figure(figsize=[12,5])
axes=fig.add_subplot(111) 
axes.boxplot(
    x=[
        tips_df[tips_df['sex']=='Female']['total_bill'],
        tips_df[tips_df['sex']=='Male']['total_bill']
        
    ],
    label=['Female','Male']
    
)
plt.show()
```

```python
fig=plt.figure(figsize=[12,5])
axes=fig.add_subplot(111) 
axes.boxplot(
    x=[
        tips_df[tips_df['smoker']=='No']['total_bill'],
        tips_df[tips_df['smoker']=='Yes']['total_bill']
        
    ],
    label=['No','Yes']
    
)
plt.show()
```

```python
fig=plt.figure(figsize=[12,5])
axes=fig.add_subplot(111) 
axes.boxplot(
    x=[
        tips_df[tips_df['day']=='Sun']['total_bill'],
        tips_df[tips_df['day']=='Sat']['total_bill']
        
    ],
    label=['Sat','Sun']
    
)
plt.show()
```

```python
fig=plt.figure(figsize=[12,5])
axes=fig.add_subplot(121) 
axes.boxplot(
    x=[
        tips_df[tips_df['time']=='Lunch']['tip'],
        tips_df[tips_df['time']=='Dinner']['tip']
        
    ],
    labels=['Lunch','Dinner'] 
)
plt.grid()
axes1=fig.add_subplot(122) 
axes1.boxplot(
    x=[
        tips_df[tips_df['size']==2]['tip'],
        tips_df[tips_df['size']==3]['tip'],
        tips_df[tips_df['size']==4]['tip']
        
    ],
    labels=['2','3','4'] 
)
plt.grid()
axes.set_xlabel('time')
axes.set_ylabel('count')
axes.set_title('Boxplot of Tips by Time')

axes1.set_xlabel('size')
axes1.set_ylabel('count')
axes1.set_title('Boxplot of Tips by Size')


plt.show()
```

# scatter, 산점도

```python
# 다변량 표현
# 성별에 따라 -크기, 색상
# 값의 크기에 따라 - 점의 크기

fig=plt.figure(figsize=[12,5])
axes=fig.add_subplot(111) 
colors = {
    'Male': 'blue',    # 남성은 파란색
    'Female': 'red'    # 여성은 빨간색
}
axes.scatter(tips_df['total_bill'], tips_df['tip'],
              s=tips_df['size']**2*10, #table size, tips_df
              c=tips_df['sex'].map(colors)
             )
plt.show()
```

```python
fig=plt.figure(figsize=[6,5])
axes=fig.add_subplot(111) 
colors = {
    'Lunch' :'skyblue', # 다른곳에서 스포이드 해오면 앞에 #붙힌다
    'Dinner' : 'pink'
}



axes.scatter(tips_df['total_bill'], tips_df['tip'],
              s=tips_df['tip']**2*10, #table size, tips_df
              c=tips_df['time'].map(colors),
              alpha=0.5 #투명도
              
             )
plt.show()
```

```python
fig=plt.figure(figsize=[6,5])
axes=fig.add_subplot(111) 
colors = {
    'Male': 'lightgreen',   
    'Female': '#ff00ad'    
}
tips_df['tip_pct']=tips_df['tip']/tips_df['total_bill']

axes.scatter(tips_df['total_bill'], tips_df['tip'],
              s=tips_df['tip_pct']*1000, #table size, tips_df
              c=tips_df['sex'].map(colors),
              alpha=0.3,
              marker='^'
             )
plt.show()
```

```python
# 파이 차트 : 비율
# 성별 비율 비교
# 빈도수 

sex_counts=tips_df['sex'].value_counts()
fig=plt.figure()
axes=fig.add_subplot(111) 
axes.pie(sex_counts,
         colors=('lightgreen','#ff00ad'),
         labels=sex_counts.index,  
         autopct='%1.1f%%',
         startangle=90)

plt.show()
```

```python
sm_counts=tips_df['smoker'].value_counts()
fig=plt.figure()
axes=fig.add_subplot(111) 
axes.pie(sm_counts,
         colors=('lightblue','lightpink'),
         labels=sm_counts.index,  
         autopct='%.2f%%', # 숫자 뒤에 %지정
         startangle=270)

plt.show()
```
