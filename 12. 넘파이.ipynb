```python
import numpy as np
data = [
    [120, 135, 150, 145],  # 제품A, 1분기~4분기
    [100, 110, 140, 130],  # 제품B
    [90, 100, 95, 105] ]
data
```

# 데이터 시뮬레이션

1차원 배열의 랜덤워크 시뮬레이션

```python
import matplotlib.pyplot as plt

n_step = 1000
step = np.random.choice([1,-1], size = n_step) #y축은 초이스에서 정해짐, 1걸음마다 플마 1
walk = np.cumsum(step)

plt.plot(walk)
```
